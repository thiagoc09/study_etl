```python
import pandas as pd
```

```python
#extração dos dados

acoes = pd.read_csv("/content/tb_acoes.csv")
anos = pd.read_csv("/content/tb_anos.csv")
programas = pd.read_csv("/content/tb_programas.csv")
orcamentos = pd.read_csv("/content/tb_orcamentos.csv")
```

```python
merge1 = orcamentos.merge(acoes,
                          left_on="fk_acao",
                          right_on="id_acao",
                          how="left")
```

```python
merge2 = merge1.merge(anos,
                          left_on="fk_ano",
                          right_on="id_ano",
                          how="left"
                          )
```

```python
merge3 = merge2.merge(programas,
                      left_on="fk_programa",
                      right_on="id_programa",
                      how="left")
```

```python
df = merge3
```

```python
df.head()
```

```python
df = df.drop(columns=["id_orcamento","fk_acao","fk_programa","fk_unidade","fk_ano","id_acao","id_ano","id_programa"])
```

```python
#Carregamento para o BigQuery

from google.oauth2 import service_account
```

```python
credentials = service_account.Credentials.from_service_account_file(filename="key.json", scopes = ["https://www.googleapis.com/auth/cloud-platform"])
```

```python
df.to_gbq(credentials=credentials,
          destination_table="curso_etl.primeiro_etl")
```
